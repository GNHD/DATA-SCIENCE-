```python
# 1. Read Total profit of all months and show it using a line plot.  

# Generated line plot must include the following properties:  

# X label name = Month Number, Y label name = Total profit 
import numpy as np
import pandas as pd
import matplotlib.pyplot as ply
df=pd.read_csv("company_sales_data.csv")
profit=df['total_profit'].tolist()
month=df['month_number'].tolist()
ply.plot(month,profit,label='Month-wise profit')
ply.xlabel('Month Number')
ply.ylabel('profit')
ply.show()
```

```python
# 2. Get total profit of all months and show line plot with the following Style properties: 


total=df['total_units'].tolist()
ply.plot(month,total,label="profit and sold items",color='r',marker='o',markerfacecolor='k',linewidth='3',linestyle='--')
ply.xlabel('Month Number')
ply.ylabel('Sold units number')
ply.legend(loc='lower right')
```

```python
# 3. Read all product sales data and show it using a multiline plot. 

# Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product). 

fashcream=df['facecream'].tolist()
fashwash=df['facewash'].tolist()
paste=df['toothpaste'].tolist()
soap=df['bathingsoap'].tolist()
shampoo=df['shampoo'].tolist()
moisturizer=df['moisturizer'].tolist()
ply.plot(month,fashcream,label='facecreams sold',linewidth='3',marker='o')
ply.plot(month,fashwash,label='facewash sold',linewidth='3',marker='o')
ply.plot(month,paste,label='toothpaste sold',linewidth='3',marker='o')
ply.plot(month,soap,label='soaps sold',linewidth='3',marker='o')
ply.plot(month,shampoo,label='shampoo sold',linewidth='3',marker='o')
ply.plot(month,moisturizer,label='moisturizer sold',linewidth='3',marker='o')
ply.xlabel('Month Number')
ply.ylabel('Sales ')
ply.legend(loc='upper left')
ply.show()
```

```python
# 4. Read toothpaste sales data of each month and show it using a scatter plot.  

# Also, add a grid in the plot. gridline style should “–“. 
paste=df['toothpaste'].tolist()
ply.scatter(month,paste,label='toothpaste sold')
ply.xlabel('Month Number')
ply.ylabel('Sales ')
ply.grid(True, linewidth= 1, linestyle="--")
ply.show()
```

```python
# 5. Read face cream and facewash product sales data and show it using the bar chart.

facecream=df['facecream'].tolist()
facewash=df['facewash'].tolist()
ply.bar([a-0.25 for a in month], facecream, width= 0.25, label = 'Face Creams sold', align='edge')
ply.bar([a+0.25 for a in month], facewash, width= -0.25, label = 'Face Wash sold', align='edge')
ply.xlabel('Month Number')
ply.ylabel('Sales ')
ply.show()
```

```python
# 6. Read sales data of bathing soap of all months and show it using a bar chart. 

# Save this plot to your hard disk. 

soap=df['bathingsoap'].tolist()
ply.bar(month,soap,label='soaps sold')
ply.xlabel('Month Number')
ply.ylabel('Sales ')
ply.savefig('Untitled Folder 5.png')
ply.show()
```

```python
# 7. Read the total profit of each month and show it using the histogram to see the most common profit ranges. 
profit=df['total_profit'].tolist()
labels=['low','average','best','worst']
profit_range = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
ply.hist(profit,profit_range,label='profit data')
ply.legend(loc='upper left')
ply.title("Profit")
ply.show()
```
