```python
#20
def corrplot(wid,hei,x,a1,a2,col,labelx,labely):
    f = plt.figure()
    f.set_figwidth(wid)
    f.set_figheight(hei)
    plt.xlabel(labelx)
    plt.ylabel(labely)
    plt.plot(x,a1,color=col)
    plt.plot(x,a2,color=col)
```
